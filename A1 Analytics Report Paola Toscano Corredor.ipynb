## Problem 2

### Step 1

After doing the EDA of the indicators file, it was necessary to make a deeper filtering in order to get the exact data that is required for the analysis. 

By using a SQL code:
 - 1 -> SELECT the important columns
 - 2 -> Filter by the Indicator Name of 'Merchandise imports from developing economies outside region (% of total merchandise imports)'
 - 3 -> Using just the information of USA and the years from 2000 to 2014 (the last year of the data)

```python
#SQL code to filter and clean more the data by county, in this case, imports in USA
with sqlite3.connect('test.db') as connection:
     pd.read_csv('World Bank/world-development-indicators/Indicators.csv').to_sql(
        name = 'indicators',
        con = connection,
        index = False,
        if_exists = 'replace')

with sqlite3.connect('test.db') as connection:
    sql1 = pd.read_sql(sql = "SELECT [CountryName],[IndicatorName],[IndicatorCode],[Year],[Value] FROM indicators WHERE [IndicatorName] = 'Merchandise imports from developing economies outside region (% of total merchandise imports)' AND [CountryCode] = 'USA' AND [Year] IN ('2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014')",
                      con = connection)
sql1
```

### Step 2

The code is performing a linear regression on the previous dataframe with columns Year and Value and fitting a curve to the data points, for the imports in the United States.

The result 5.668465191465208e-22 0.026238159105836198 are the values of alpha and beta respectively. The value of beta is 0.026238159105836198, which means that for every increase in Year by one unit, the value of Value increases by 0.026238159105836198. 

It can be seen that imports have had a positive trend over time.

```python
#Graph that shows the growth in imports over the years
imports_graph = sql1
X,Y = imports_graph[['Year','Value']].values.T

yp = np.log(Y)
n = len(X)
K = np.array([[n,np.sum(X)],[np.sum(X),np.sum(X**2)]])
F = np.array([np.sum(yp),np.sum(X*yp)])

lnalpha,beta = np.linalg.solve(K,F)
alpha = np.exp(lnalpha)
print(alpha,beta)

_X = np.linspace(2000,2023)
_Y = alpha*np.exp(_X*beta)

plt.plot(X,Y,'-', color='blue', linewidth=4)
plt.plot(_X,_Y,'--', color='red', linewidth=4) 
plt.grid()
plt.xlabel("Year")
plt.ylabel("Merchandise imports")
plt.gca().set_yscale('log')
plt.show()
```
